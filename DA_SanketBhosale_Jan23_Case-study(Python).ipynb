# Create a Shoping cart with the following basic functionality

```python

stock={'1':['Biscuits: a pack of 20',5,2.50],
      '2':['Cereals: 2 pounds/pack',10,3.50],
      '3':['Chicken: 1 pound/pack',20,13.30],
      '4':['Oats: 4 pounds/pack',20,3.90],
      '5':['Rice: 2 pounds/pack',30,4.50]}
```

```python
# Display the list of items avilable in store along with the quantity and cost.
def display():
    value=stock.values()
    print('*''%55s'%('ESSENTIAL ITEMS AVAILABLE'))
    print()
    print('%20s %20s %20s %15s'%('Sr.no','Items','Quantity','Cost/Item'))
    for code,item in stock.items():
        print('%17s %30s %10s %15s'%(code,item[0],item[1],item[2]))
    print()
# Ask the costomer to enter the number of quantity required for each item
    biscuit=float(input('Enter the quantity of biscuits:'))
    stock['1'][1]-biscuit
    cereal=float(input('Enter the quantity of Cereals:'))
    stock['2'][1]-cereal
    chicken=float(input('Enter the quantity of Chicken:'))
    stock['3'][1]-chicken
    oats=float(input('Enter the quantity of Oats:'))
    stock['4'][1]-oats
    rice=float(input('Enter the quantity of Rice:'))
    stock['5'][1]-rice
# Ask the customer for name, address and distance from the store.
    print()
    name=input('Enter Your Name: ')   # Ask customer name
    ad=input('Enter Your Address :')  # Ask for a Address
    dist=int(input('Enter Distance from Store:')) #Delivery Charge chart 
# Delivery charges as per the distance from the store
    if dist<=5:
        delivery_charge=10
    elif dist<=10:
        delivery_charge=12
    elif dist<=15:
        delivery_charge=14
    else:
        delivery_charge=20
    print()
    print('*''%55s'%('BILL')) # Billing process starts
    print()
    print('%20s %20s %20s %15s'%('Sr.no','Items','Quantity','Cost/Item'))  # bill in tabular format
    
    print('%17s %30s %10s %15s'%(1,'Biscuits',biscuit,biscuit*2.5))
    print('%17s %30s %10s %15s'%(2,'Cereal',cereal,cereal*3.5))
    print('%17s %30s %10s %15s'%(3,'Chicken',chicken,chicken*13.3))
    print('%17s %30s %10s %15s'%(4,'Oats',oats,oats*3.90))
    print('%17s %30s %10s %15s'%(5,'Rice',rice,rice*4.5))
    print()
# calculating total bill
    total_cost=biscuit*2.5+cereal*3.5+chicken*13.3+oats*3.90+rice*4.5   
    print('total amount',total_cost)
    print('Delivery charges',delivery_charge)
    print('Grand Toal',total_cost+delivery_charge)
    print('Name :',name)
    print('Address :',ad)
    stock['1'][1]=stock['1'][1]-biscuit   # update the value
    stock['2'][1]=stock['2'][1]-cereal
    stock['3'][1]=stock['3'][1]-chicken
    stock['4'][1]=stock['4'][1]-oats
    stock['5'][1]=stock['5'][1]-rice
    loop1=input('Do you want to continue shopping?(yes/no)') # loop for shopping
    if loop1=='no':
        return False
    else:
        print('yes')
        return True
while True:
    if not display():
        break
    
    print('Thank You:Visit Again') # break the code
```
